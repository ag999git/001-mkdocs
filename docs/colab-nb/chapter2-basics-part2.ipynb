```python
my_str = 'Hello World!' # A string assigned to variable my_str
print(str(my_str))  # Expected output: Hello World!
# str() converts to a string (although my_str is already a string)

print(my_str[0])  # Expected output: H
# Indexing starts from 0, so my_str[0] gives the first character.

print(my_str[1:5])  # Expected output: ello
# Characters from index 1 to index 4 (5 is excluded).
# Slice length = 5 - 1 = 4 characters.

print(my_str + ' From me')  # Expected output: Hello World! From me
# + joins two strings together.

print(my_str * 3)  # Expected output: Hello World!Hello World!Hello World!
# * repeats the string 3 times.

print(len(my_str))  # Expected output: 12
# len() returns the number of characters in the string.

print(my_str[-1])  # Expected output: !
# Negative indexing: -1 refers to the last character.

print(my_str.upper())  # Expected output: HELLO WORLD!
# .upper() converts all letters to uppercase.

print(my_str.lower())  # Expected output: hello world!
# .lower() converts all letters to lowercase.

print('World' in my_str)  # Expected output: True
# 'in' checks whether the substring exists inside the string.

print(my_str.replace('World', 'Python'))  # Expected output: Hello Python!
# replace(old, new) substitutes part of the string.
```

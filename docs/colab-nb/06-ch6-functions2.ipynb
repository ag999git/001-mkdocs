The following script shows how Python handles different object types during function call.

```python
def process_data(z):
    """
    Shows how Python handles different object types during function call.
    """
    if isinstance(z, int):  # Check if z is an integer (immutable)
        print(f"--- Processing Integer (Immutable) ---")
        print(f"Initial internal z: {z} (ID: {id(z)})")

        # This does NOT change the original 'a'.
        # It creates a new local variable 'z' pointing to 4.
        z = 4
        print(f"Modified internal z: {z} (ID: {id(z)})")

    elif isinstance(z, list):  # Check if z is a list (mutable)
        print(f"\n--- Processing List (Mutable) ---")
        print(f"Initial internal z: {z} (ID: {id(z)})")

        # This modifies the object in-place.
        # Both 'c' and 'z' point to the same memory ID.
        z.append(1)
        print(f"Modified internal z: {z} (ID: {id(z)})")

# --- Main Execution ---

# 1. Testing with an Integer
a = 5
print(f"Original 'a' before call: {a} (ID: {id(a)})")
process_data(a)
print(f"Original 'a' after call: {a} (Still 5!)")  # Out (5)

# 2. Testing with a List
c = [5, 4, 3]
print(f"\nOriginal 'c' before call: {c} (ID: {id(c)})")
process_data(c)
print(f"Original 'c' after call {c} (Updated to include 1!)")  #Out([5, 4, 3, 1])
```

#### Default Arguments (Default Values in Function Definition)
In Python, every function appears in two places: (1) Where it is defined (function definition) and (2) Where it is used (function call). Normally, when you call a function, you must provide values (called arguments) that match the parameters given in the function definition.
But sometimes, you have a situation like this:
•	A function needs a parameter whose value is usually the same every time.
•	Only occasionally does the user want to override that value.
In such cases, Python allows you to use default arguments. It is possible to provide default values to a function parameter in the function definition. This makes programming in Python very easy. Default values are always provided in the function definition, not in the function call.
How it works:- (1) If a default value is provided in a function definition, then you have the option of providing a value or not providing a value in a function call. (2) If you provide a value to the argument in the function call then this value will be taken, but if you do not provide a value, then the default value is taken. The following script shows the use of default arguments:-


```python
# Example demonstrating DEFAULT ARGUMENTS in Python

# Function definition with default arguments
def greet(name="Student", greeting="Hello"):
    # Print the greeting message
    print(greeting, name)

# Testing the function with different combinations of arguments

# 1. Both arguments supplied. Parameter 'name' takes supplied value "Alice"
# and parameter 'greeting' takes supplied value "Good morning".
# Default values are NOT used
greet("Alice", "Good morning")
# Output: Good morning Alice

# 2. Only the first argument supplied. Parameter 'name' takes supplied value "Bob".
# Parameter 'greeting' takes the default value "Hello" since it is not supplied.
greet("Bob") # Output: Hello Bob

# 3. No arguments supplied. Parameter 'name' takes default value "Student"
# Parameter 'greeting' takes default value "Hello". Both default values are used.
greet() # Output: Hello Student
```

Example: Problem with Mutable Default Argument

```python
# Example demonstrating the problem with mutable default arguments

def add_number_to_list(number, number_list=[]):
    # Adds a number to a list and returns the list.
    # Warning: Using mutable default argument (eg list) can lead to unexpected results
    # The default list is created only once when the function is defined, not each
    # time the function is called. So, if you modify this list (like appending a number)
    # it will affect all future calls that use the default list.

    # Add the number to the list
    number_list.append(number)

    # Return the modified list
    return number_list

# Testing the function wich has mutable default argument
result1 = add_number_to_list(2)
print("After first call:", result1)     # Expected: [2]

result2 = add_number_to_list(3)
print("After second call:", result2)    # Should be 3 but is Unexpected: [2, 3]
```

The recommended solution is to use None as the default value and create the mutable object inside the function.
Exampl is as below



```python
def add_number_to_list(number, number_list=None):
    # Correct implementation. Create a new list if no list was provided.
    # This ensures that each call to the function gets its own separate list.
    if number_list is None:
        number_list = []

    # Add the number to the list
    number_list.append(number)

    return number_list

# Testing the corrected function
print(add_number_to_list(2))   # [2] in 1st call
print(add_number_to_list(3))   # [3] in 2nd call, not affected by the first call due to the use of None as default value.
```

**Passing Arguments by Position (Positional Arguments)**
In positional arguments, the values in the function call are matched with parameters based on their order. The first argument goes to the first parameter, the second argument to the second parameter, and so on.


```python
# Example demonstrating POSITIONAL ARGUMENTS in Python

# Function definition
def display_water_sources(river, lake, sea):
    # Displays three types of water bodies. When positional arguments are used,
    # values are matched to parameters strictly by their position.
    print("River:", river, "Lake :", lake, "Sea  :", sea)

# RULE 1 : ORDER MATTERS
# Correct order
display_water_sources("Ganga", "Dal Lake", "Arabian Sea")
# Output: River: Ganga, Lake : Dal Lake, Sea  : Arabian Sea

# Incorrect logical order but valid syntax). Program runs but meaning is wrong
display_water_sources("Arabian Sea", "Ganga", "Dal Lake")
# Output: River: Arabian Sea, Lake : Ganga, Sea  : Dal Lake.

# RULE 2 : NUMBER OF ARGUMENTS MUST MATCH PARAMETERS
# Correct: exactly 3 arguments
display_water_sources("Nile", "Victoria", "Mediterranean")

# ERROR: Missing one argument. Only river and lake provided, sea is missing.
# display_water_sources("Nile", "Victoria"). # Error if uncommented
# TypeError: missing 1 required positional argument: 'sea'

# ERROR: Too many arguments. Four provided but only three parameters defined.
# display_water_sources("Nile", "Victoria", "Mediterranean", "Pacific")
# TypeError: takes 3 positional arguments but 4 were given

# RULE 3 : POSITIONAL ARGUMENTS WITH DEFAULT VALUES
def display_water_info(river, lake="Unknown Lake", sea="Unknown Sea"):
    # Example showing positional arguments with default values.
    print("River:", river, "Lake :", lake, "Sea  :", sea)

# Only river provided → others use default values
display_water_info("Ganga")  # River: Ganga, Lake : Unknown Lake, Sea  : Unknown Sea

# River and lake provided
display_water_info("Ganga", "Chilika")  # River: Ganga, Lake : Chilika, Sea  : Unknown Sea

# All values provided
display_water_info("Ganga", "Chilika", "Bay of Bengal")  # River: Ganga, Lake : Chilika, Sea  : Bay of Bengal
```

```python
# Demonstration of KEYWORD ARGUMENTS in Python

# Function definition
def student_info(name, age, course):
    # Displays information about a student.
    # Parameters: name-> student name, age-> student age, course-> course enrolled
    print("Name:", name, "Age:", age, "Course:", course)

# RULE 1 : ORDER DOES NOT MATTER
# Arguments are matched by parameter name, not by position
student_info(age=20, course="Physics", name="Ravi")  # OK
student_info(course="Mathematics", name="Anita", age=21)  # Still OK

# RULE 2 : PARAMETER NAMES MUST MATCH EXACTLY
# Correct example
student_info(name="Arjun", age=22, course="Chemistry")

# Incorrect example (commented out)
# The parameter 'subject' does not exist in function definition. Should be 'course'.
# student_info(name="Ravi", age=19, subject="Biology")  # subject not valid parameter
# TypeError: student_info() got an unexpected keyword argument 'subject'

# RULE 3 : A KEYWORD ARGUMENT CANNOT BE REPEATED
# Incorrect example (commented out)
# student_info(name="Ravi", age=20, name="Amit")  # name is repeated
# SyntaxError: keyword argument repeated

# RULE 4 : POSITIONAL ARGUMENTS MUST COME FIRST
# Correct example
# Positional argument first, then keyword arguments

student_info("Kiran", age=23, course="Economics")

# Incorrect example (commented out)
# student_info(name="Ravi", 20, "Physics")  # Positional argument (20) cannot follow keyword argument (name="Ravi")
# SyntaxError: positional argument follows keyword argument

# Another incorrect example

# student_info("Ravi", age=20, "Physics")
# This would cause:
# SyntaxError: positional argument follows keyword argument

# SUMMARY

# Keyword arguments must match parameters using names
# Order does not matter
# Parameter names must match exactly
# A keyword cannot be repeated
# Positional arguments must come before keyword arguments
```

The following script shows the use of variable Number of Arguments (*args) for adding numbers. The numbers are passed as arguments in the function call and are collected and stored in a tuple named args.

```python
# Function that accepts variable number of arguments
def add_numbers(*values):
    # We are using *values. We can any name. *args is common but not required
    # The * tells Python to collect all additional positional arguments into
    # a tuple named 'values'. values will be empty if no arguments provided.

    total = 0  # Start with zero. We will add all numbers in values to this total

    # Iterate through the tuple and add each value to the total
    for value in values:
        total = total + value

    # Return the result as a f-string that includes tuple values and total
    return f"Got, {values}. Total is {total}"

# Function calls with different numbers of arguments
print(add_numbers(2, 3))          # Got, (2, 3). Total is 5
print(add_numbers(1, 2, 3, 4))    # Got, (1, 2, 3, 4). Total is 10
print(add_numbers(10))            # Got, (10,). Total is 10
print(add_numbers())              # Got, (). Total is 0
```

6.13.	Combining Fixed and Variable Arguments
You can combine fixed parameters with *args. The following script takes the first argument passed into the parameter power and all remaining into the tuple called Example: Raise numbers to a power and add them.


```python
# Function that raises numbers to a given power and sums them
# The first parameter (power) is the power to which numbers will be raised.
# The rest of the parameters (*values) are the numbers to be processed.
# We use *values to accept any number of numbers.
def sum_of_powers(power, *values):

    total = 0
    my_str = ""
    # Each number is raised to the specified power
    for value in values:
        total = total + (value ** power)
        # my_str is built up to show the expression being calculated. For example
        # if power=2 and values=(1,2,3), my_str is "1**2 + 2**2 + 3**2 + "
        my_str = my_str + str(value) + " ** " + str(power) + " + "

    my_str = my_str.rstrip(" + ")  # Remove the trailing " + " from the expression string
    return f"Power = {power}, Numbers = {values}, Expression → {my_str} = {total}"

# Example calls
print(sum_of_powers(2, 1, 2, 3))   # Power = 2, Numbers = (1, 2, 3), Expression → 1**2 + 2**2 + 3**2 = 14
print(sum_of_powers(3, 2, 3))      # Power = 3, Numbers = (2, 3), Expression → 2**3 + 3**3 = 35
```

**Explicit Argument Enforcement (/ and *)**
  
Since Python 3.8, we can use two special symbols in a function header to dictate exactly how arguments must be passed:
1.	The / (Slash): Everything to the left of this symbol must be Positional-only.
2.	The * (Asterisk): Everything to the right of this symbol must be Keyword-only.


```python
# PART 1: POSITIONAL-ONLY ARGUMENTS (The / Marker)

def calculate_area(length, width, /):
    """Everything before '/' must be passed without a name."""
    return length * width

# --- Testing Part 1 ---
# VALID CALLS
print(calculate_area(10, 5))          # Simple positional call
print(calculate_area(15.5, 2))        # Positional with float

# INVALID CALLS (Will raise TypeError)
# calculate_area(length=10, width=5)  # Error: Cannot use names
# calculate_area(10, width=5)         # Error: Even one name is forbidden

# PART 2: KEYWORD-ONLY ARGUMENTS (The * Marker)

def print_result(*, unit, label="Area"):
    """Everything after '*' must be passed with a name."""
    print(f"{label}: {unit}")

# --- Testing Part 2 ---
# VALID CALLS
print_result(unit="sq cm")            # Required keyword provided
print_result(label="Size", unit="m")  # Keywords can be in any order

# INVALID CALLS (Will raise TypeError)
# print_result("cm")                  # Error: Must specify 'unit='
# print_result("Area", "cm")          # Error: Positional values forbidden
# print_result()                      # Error: Missing required keyword 'unit'

# PART 3: COMBINED ENFORCEMENT (Both / and *)

def get_shape_info(length, width, /, *, unit, color="Blue"):
    """
    length, width: Must be positional (before /)
    unit, color:   Must be keywords (after *)
    """
    area = length * width
    print(f"Shape ({color}) -> Area: {area} {unit}")

# --- Testing Part 3 ---
# VALID CALLS
get_shape_info(10, 5, unit="m2")                # 2 positional, 1 keyword
get_shape_info(8, 4, color="Red", unit="cm")    # Keyword order is flexible

# INVALID CALLS (Will raise TypeError)
# get_shape_info(length=10, width=5, unit="m2") # Error: / forbids use of names for length/width
# get_shape_info(10, 5, "m2")            # Error: * blocks positional for unit
# get_shape_info(10, 5, "m2", "Red")     # Error: Too many positional values
```

Example of recursive function to calculate factorial of a positive integer

```python
# Factorial using Recursion. n! = n × (n-1) × (n-2) × ... × 2 × 1

def factorial_recursive(n):

    # ---------- BASE CASE ----------
    # When n is 0 or 1, the factorial is 1
    # This stops further recursive calls.
    if n == 0 or n == 1:
        print(f"Base case reached: factorial({n}) = 1")
        return 1

    # ---------- RECURSIVE CASE ----------
    # The function calls itself with a smaller value (n-1)
    print(f"Computing: {n} × factorial({n-1})")
    result = n * factorial_recursive(n - 1)  # Recursive call.

    # Show the value being returned after recursion unwinds
    print(f"Returning: {n}! = {result}")
    return result  # Return the computed factorial value to the caller

# Main Program. Ask the user for input. We are not validating
# input here. Assume user enters a valid non-negative integer
number = int(input("Enter a non-negative integer: "))
# Call the recursive function
factorial_value = factorial_recursive(number)
# Display the final result
print(f"\nFinal Result: {number}! = {factorial_value}")

```

Recursive function to generate Fibonacci numbers

```python
# Recursive function to compute the nth Fibonacci number

def fibonacci_recursive(n):
    # Returns the nth Fibonacci number using recursion.
    # There are 2 base cases: Fibonacci(0) = 0 and Fibonacci(1) = 1

    # -------- Base Case 1 --------
    # If n = 0, return
    if n == 0:
        return 0

    # -------- Base Case 2 --------
    # If n = 1, return 1
    elif n == 1:
        return 1

    # -------- Recursive Case --------
    # Compute the sum of the previous two Fibonacci numbers
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Test calls

print(f"Fibonacci(5)  → {fibonacci_recursive(5)}")   # 5
print(f"Fibonacci(6)  → {fibonacci_recursive(6)}")   # 8
print(f"Fibonacci(7)  → {fibonacci_recursive(7)}")   # 13
print(f"Fibonacci(10) → {fibonacci_recursive(10)}")  # 55
```

Tower of Hanoi

```python
# Tower of Hanoi using three lists representing towers A, B, C. Each list acts
# like a stack. In the list leftmost item is at bottom (Biggest) and rightmost
# is at top (Smallest). The script also counts and displays move numbers.

def print_towers(towers):
    """Display the current state of all three towers."""
    print(f"A: {towers['A']}   B: {towers['B']}   C: {towers['C']}")
    print("-" * 40)

def move_disk(towers, source, target, move_counter):
    # Move top disk from one tower to another also increment the move counter

    disk = towers[source].pop()     # Remove top disk from source
    towers[target].append(disk)     # Place disk on target

    move_counter[0] += 1            # Increment move count

    print(f"Move {move_counter[0]}: disk from {source} → {target}")
    print_towers(towers)

def hanoi_recursive(n, source, target, auxiliary, towers, move_counter):
    """
    Recursive solution to Tower of Hanoi.

    n         -> number of disks
    source    -> source tower (A/B/C)
    target    -> target tower (A/B/C)
    auxiliary -> auxiliary tower (A/B/C)
    towers    -> dictionary storing the towers
    move_counter -> list used to track number of moves
    """

    # -------- Base Case --------
    if n == 1:
        move_disk(towers, source, target, move_counter)
        return

    # -------- Recursive Case --------

    # Step 1: Move n-1 disks from source → auxiliary
    hanoi_recursive(n - 1, source, auxiliary, target, towers, move_counter)

    # Step 2: Move largest disk from source → target
    move_disk(towers, source, target, move_counter)

    # Step 3: Move n-1 disks from auxiliary → target
    hanoi_recursive(n - 1, auxiliary, target, source, towers, move_counter)

# INITIAL SETUP

num_disks = 3

# Towers represented as a dictionary. Create initial state for tower A with
# disks from largest on left to smallest on right.
initial_state_A = list(range(num_disks, 0, -1))
# Initialize towers with initial_state_A for tower A. Empty lists for B & C
towers = {"A": initial_state_A, "B": [], "C": []}

# Move counter stored in a list so recursion can update it
move_counter = [0]

print("Initial State")
print_towers(towers)

# Solve the puzzle
hanoi_recursive(num_disks, "A", "C", "B", towers, move_counter)

# Display total moves
print(f"Total moves required: {move_counter[0]}")
```

Fibonacci using Memoization (Optimized Recursion)

```python
# Fibonacci using Memoization (Optimized Recursion)

memo = {0: 0, 1: 1}
# memo is a dictionary that will store previously computed Fibonacci values.
# It starts with the base cases: F(0) = 0 and F(1) = 1. As the recursive
# function computes new Fibonacci numbers, it will add them to this dictionary,
# allowing for fast retrieval in future calls.

def fib_memo(n):
    # Returns the nth Fibonacci number using memoization to optimize recursion.
    # It checks if value is already computed & stored in the memo dictionary.
    # If it is, it returns that value immediately. If not, it computes it
    # recursively, stores the result in the memo, and then returns it.

    # -------- Step 1: Check if value already exists in cache --------
    # If yes, return it immediately (very fast)
    if n in memo:
        return memo[n]

    # -------- Step 2: Otherwise compute recursively --------
    # Break the problem into two smaller Fibonacci problems
    fib_n_minus_1 = fib_memo(n - 1)
    fib_n_minus_2 = fib_memo(n - 2)

    result = fib_n_minus_1 + fib_n_minus_2

    # -------- Step 3: Store the result in the cache --------
    memo[n] = result

    # -------- Step 4: Return the computed value --------
    return result

# TEST CASES
print("Fibonacci(10):", fib_memo(10))   # Expected: 55
print("Fibonacci(15):", fib_memo(15))   # Expected: 610
print("Fibonacci(20):", fib_memo(20))   # Expected: 6765

# Display the memoization dictionary
# This shows all Fibonacci values that were stored during execution
print("Memoization Cache Contents:", memo)
```

The following script gives a Simple Generator Example

```python
# Simple generator example

def simple_generator():
    yield 1  # yield used to produce a value and pause function's execution
    yield 2  # yield produces the next value when generator is resumed
    yield 3  # yield produces final value after which generator is exhausted

# Create a generator object
gen = simple_generator()

# Get values one by one
print(next(gen))   # 1
print(next(gen))   # 2
print(next(gen))   # 3
```

Using a Generator with a for-loop. Generators are most commonly used with for-loops. The following script shows use of generators in for loop:-

```python
# Using generator in a for loop

def simple_generator():
    yield "A"  # Produces value "A". Pauses until next value requested
    yield "B"  # Produces value "B" when the generator is resumed
    yield "C"  # Produces value "C". After this generator is exhausted

for value in simple_generator(): # Retrieves values from generator till finish
    print(value)
```

Infinite counter generator

```python
# Infinite counter generator

def infinite_counter():
    num = 1
    while True:  # Loop indefinitely
        yield num  # Yield current number. Pause until next value requested
        num += 1

counter = infinite_counter()
print(next(counter))   # 1
print(next(counter))   # 2
print(next(counter))   # 3

```

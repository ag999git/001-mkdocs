The following script shows the use of some of the file object/ handle methods/ functions discussed in the book

```python
# Demonstrating file (1) pointer manipulation (2) reading (3) closing
# 1. Create and write some content first (so that reading works)
with open("demo.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3\n")  # Write multiple lines to the file

# 2. Now open the file for reading
with open("demo.txt", "r") as f:
    print("Initial pointer position:", f.tell())   # Output: 0

    # A. read() - read first 6 characters
    print("Using read(6):", f.read(6))  # Output: 'Line 1'
    print("Pointer after read:", f.tell())         # Output: 6

    # B. seek() - move pointer back to beginning
    f.seek(0)
    print("Pointer after seek(0):", f.tell())  # Output: 0

    # C. readline() - read one line
    print("Using readline():", f.readline())  # Output: 'Line 1\n'

    # D. readlines() - read remaining lines as list
    print("Using readlines():", f.readlines()) # Output: ['Line 2\n', 'Line 3\n']

# 3. File is automatically closed after 'with'
print("File closed automatically by with statement.")

# 4. Demonstrating close() explicitly (not recommended when using 'with')
f = open("demo.txt", "r")  # open file again
print("File opened again. Closed?", f.closed)  # Output: False

f.close()   # close()
print("After close(), Closed?", f.closed)  # Output: True
```

The following script shows the use of write methods:

```python
# Open file in write mode. 'w' will create the file if it does not exist
# and overwrite (delete) existing content if it exists
with open("write_demo.txt", "w") as f:

    # A. write() → writes a single string
    # Note: newline (\n) must be added manually
    f.write("Line 1\n")  # Writes 'Line 1' followed by a newline character
    f.write("Line 2\n")  # Writes 'Line 2' followed by a newline character

    # B. writelines() → writes multiple strings from a list
    # Note: it does NOT add newline automatically
    lines = ["Line 3\n", "Line 4\n"]  # List of strings to write
    f.writelines(lines)  # Writes each string from the list named "lines"

# File is automatically closed here due to 'with'
```

The following script shows the use of various file object attributes:

```python
# 1. Use 'w' mode to create or overwrite file, & specify encoding for text files
with open("attr_demo.txt", "w", encoding="utf-8") as f:
    f.write("Hello\nWorld\n")

# 2. Open again for reading
f = open("attr_demo.txt", "r", encoding="utf-8")

# 3. Basic attributes
print("File name:", f.name)        # attr_demo.txt
print("Mode:", f.mode)             # r
print("Is closed?:", f.closed)     # False

# 4. Encoding-related attribute
print("Encoding:", f.encoding)     # utf-8

# 5. Read some content to activate newline detection
f.readline()

# 6. Newline attribute (may show '\n' or None)
print("Newlines encountered:", f.newlines)  # '\n' or None if not encountered

# 7. Advanced attributes (for awareness)
print("Error handling:", f.errors) # strict (default for encoding issues)
print("Buffer object:", f.buffer)  # <_io.BufferedReader name='attr_demo.txt'>

# Close file
f.close()

# Check closed status again
print("After closing, is closed?:", f.closed)
```

The following script shows the use of above methods of the os module:

```python
import os

# STEP 1: Check current directory
cwd = os.getcwd()
print("Current Directory:", cwd)  # Current Directory: <your_current_directory>

# STEP 2: Create a new directory
os.makedirs("test_dir", exist_ok=True)
print("Directory 'test_dir' created or already exists.")

# STEP 3: Create a file inside it
# Construct file path using os.path.join for better compatibility across OS
file_path = os.path.join("test_dir", "sample.txt")
with open(file_path, "w") as f:
    f.write("Hello OS Module")
print("File 'sample.txt' (1) opened OR (2) created & opened in 'test_dir'")

# STEP 4: Check file existence
if os.path.exists(file_path):
    print("File exists")

# STEP 5: Rename file
new_path = os.path.join("test_dir", "renamed.txt")
os.rename(file_path, new_path)
print("File renamed to 'renamed.txt'.")

# STEP 6: List contents
print("Directory contents:", os.listdir("test_dir"))

# STEP 7: Delete file safely
if os.path.isfile(new_path):
    os.remove(new_path)
    print("File 'renamed.txt' deleted.")

# STEP 8: Remove directory
os.rmdir("test_dir")
print("Directory 'test_dir' deleted.")
```

 The following script shows the usage of the three standard input/output streams:

```python
import sys

# 1. Writing to stdout
sys.stdout.write("Enter your name: ")  # Prompt user for input (this will appear in the console)

# 2. Reading from stdin
name = sys.stdin.readline().strip()  # Read user input and remove any trailing newline

# 3. Writing normal output
sys.stdout.write("\nHello " + name + "\n")  # Greet the user with their name

# 4. Writing error message
sys.stderr.write("This is just a demo error message\n")  # Write an error message to stderr (appears in console, but can be redirected separately)
```

The following script shows how pickling is done using the pickle module:

```python
import pickle

# STEP 1: Create an object (dictionary)
# Object we want to serialize (convert to byte stream) & store in file.
data = {'a': 1, 'b': 2}

# STEP 2: Pickle (write object to file)
# Open a file in binary write mode to store the pickled data
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)   # Convert object → byte stream and store

# STEP 3: Unpickle (read object from file)
# Open the file in binary read mode to read the pickled data
with open("data.pkl", "rb") as f:
    new_data = pickle.load(f)  # Convert byte stream → object

# STEP 4: Verify
print("Original:", data)  # Print the original dictionary
# Output: Original: {'a': 1, 'b': 2}
print("Unpickled:", new_data)  # Print unpickled dictionary (matches original)
# Output: Unpickled: {'a': 1, 'b': 2}

# STEP 5: Reading raw file (not useful)
# Open the file in binary read mode to read the raw bytes
with open("data.pkl", "rb") as f:
    print("Raw bytes:", f.read())  # Not human-readable
    # Ouput: b'\x80\x04\x95\x0c\x00\x00\... (binary data)
```

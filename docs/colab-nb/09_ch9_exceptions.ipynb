The Basic Structure of a typical try-except block is as follows:

```python

try:
    # Risky code
except SomeException:
    # Code to handle error
else:
    # Runs only if no exception occurs

```

The following script shows the flow of execution in try-except-else-finally blocks:  

````python
def process_numbers(number, divisor):
    try:
        print("1. Performing division...")
        result = number / divisor # may raise ZeroDivisionError or TypeError

    except ZeroDivisionError:
        print(f"Error: Cannot divide by zero (divisor was {divisor})")

    else:
        print("No exception occurred")
        print("Result:", result)

    finally:
        print("Finally block always executes")

# --------- Test Cases ---------

print("Case 1: No Error")
process_numbers(10, 2)
# Case 1 Output: try → else → finally

print("Case 2: Handled Exception")
process_numbers(10, 0)  # ZeroDivisionError → handled by specific except block
# Case 2 Output: try → except(ZeroDivisionError) → finally

print("Case 3: Unhandled Exception")
process_numbers("10", 2)   # TypeError. Trying to divide a string by integer.
# Raises a TypeError which is not specifically handled in the except blocks,
# so it will remain unhandled and will show the traceback error message.
# The finally block will still execute.
# Case 3 Output: try → finally (No else, no except) → traceback error message for TypeError
````

Example Script uses "generic" error to catch all errors

````python
def process_numbers(number, divisor):
    try:
        print("1. Performing division...")
        result = number / divisor  # may cause TypeError or ZeroDivisionError

    except ZeroDivisionError:
        print( f"divisor = {divisor} Error. Cannot divide by zero.")

    except Exception as e:  # Generic except for other exceptions like TypeError
        print("Generic Error:", e)

    else:  # Runs if no exception occurs
        print("No exception occurred")
        print("Result:", result)

    finally:  # Always runs, use for cleanup
        print(f"Finally block always executes. (divisor was {divisor})")

# Case 1: No Error
print("Case 1: No Error")
process_numbers(10, 2)  # No error → result = 5.0
# Output: try → else → finally

# Case 2: ZeroDivisionError.
print("Case 2: ZeroDivisionError")
process_numbers(10, 0)  # ZeroDivisionError → because divisor is zero
# Output: except(ZeroDivisionError) → finally

# Case 3: TypeError not specifically handled so Generic except will catch it
print("Case 3: TypeError")
process_numbers(10, "a")  # TypeError → because divisor is not a number
# Output: except(generic) → finally

````

How Context Manager Handles Exceptions: A context manager ensures that cleanup code runs even if an exception occurs. Internally, it works like as shown in script below:

````python
manager = open(fileName, mode)
# open() returns file object as a context manager
# The with statement automatically calls manager.__enter__() to get the file
# object and assigns it to f. It also ensures that manager.__exit__() is called
# to close the file, even if an exception occurs.
f = manager.__enter__()

try:
    content = f.read()
except:
    # Exception handling (outside)
    raise
finally:
    manager.manager.__exit__(exc_type, exc_value, traceback)
    # exc_type, exc_value, traceback are the exception details if an exception
    # occurred, otherwise they are None.
````

The following script shows use of try-except-else-finally blocks for file reading operations:

````python
# This script demonstrates use of try-except-else-finally blocks for handling
# file operations. It has except blocks for errors like FileNotFoundError &
# PermissionError. It also has a generic except block to catch any other
# unforeseen exceptions. Using open() in 'r' mode is safer for reading
# files as it does not allow modifications, which can help prevent certain
# types of errors. The else block executes if no exceptions occur, and the
# finally block always executes regardless of whether an exception was raised
# or not, making it ideal for cleanup actions.

def myReader(fileName, mode='r'):
    try:
        print(f"1. Trying to open file in '{mode}' mode...")

        with open(fileName, mode) as f:
            content = f.read()

    except FileNotFoundError:
        print("Error: File not found")

    except PermissionError:
        print("Error: Permission denied")

    except Exception as e:
        print("Generic error - not caught by specific except blocks):", e)

    else:
        print("2. File read successfully")
        print("Content:\n", content)

    finally:
        print("3. File operation attempted")

# Test cases
print("Case 1: Valid file")
myReader("existing_file.txt")  # Make sure this file exists for the test
# Case 1 Output: try → else → finally. (No exceptions occur)

print("Case 2: File not found")
myReader("non_existent_file.txt")  # This file does not exist
# Case 2 Output: try → except(FileNotFoundError) → finally

print("Case 3: Permission denied")
myReader("/root/secret_file.txt")  # Assume file exists but not accessible
# Case 3 Output: try → except(PermissionError) → finally

print("Case 4: Invalid mode")
myReader("existing_file.txt", mode='non_existent_mode')  # Invalid mode → ValueError
# Case 4 Output: try → except(non_existent_mode) → finally
````

The following script has two user derived exceptions named ParentException and ChildException. ChildException is subclass of ParentException. If you catch ParentException before ChildException, then the ParentException will also catch the ChildException.

````python
# Step 1: Define custom exceptions
class ParentError(Exception):
    pass

class ChildError(ParentError):   # ChildError inherits from ParentError
    pass

# Case 1: Catch Child before Parent (Correct Order)
print("Case 1: Catch ChildError before ParentError")
for x in [ParentError, ChildError]:
    try:
        raise x()   # Raise ParentError or ChildError

    except ChildError:  # This will catch ChildError but not ParentError
        print("Caught ChildError")

    except ParentError: # ChildError already caught so only catches ParentError
        print("Caught ParentError")

# Case 2: Catch Parent before Child (Incorrect Order)
print("Case 2: Catch ParentError before ChildError")

for x in [ParentError, ChildError]:
    try:
        raise x()

    except ParentError:   # This will catch BOTH ParentError and ChildError
        print("Caught ParentError")

    except ChildError: # This NEVER executes ParentError catches both exceptions
        print("Caught ChildError")
````

Role of `__str__()` in Exceptions
`__str__()` is a special method (dunder method). It defines how an object is converted to a string. It is used by: (1) print(obj) and (2) str(obj)
The following script shows 2 things: (1) Use of user defined exception (2) Use of __str__() to "display" the exception. This connects two important ideas: raising an exception and how it is displayed.


````python
class InsufficientBalanceError(Exception):
    def __init__(self, balance, withdraw):
        self.balance = balance
        self.withdraw = withdraw

    def __str__(self):  # Custom error message when the exception is printed
        return f"Balance {self.balance} is less than withdrawal {self.withdraw}"

# Function to simulate money withdrawal & show custom exception handling
def withdraw_money(balance, amount):
    try:
        if amount > balance:
            raise InsufficientBalanceError(balance, amount)

        print("Withdrawal successful")

    except InsufficientBalanceError as e:  # Handle custom exception
        print("Error:", e)  #Print __str__() of InsufficientBalanceError class

# Test cases
print("Case 1: Sufficient balance")
withdraw_money(1000, 500)  # Balance is sufficient → Withdrawal successful

print("Case 2: Insufficient balance")
withdraw_money(1000, 1500)  # Balance is insufficient
# Error: Balance 1000 is less than withdrawal 1500
````

The following script shows how to access the exception object:

````python
# Demonstrating Exception Object
try:
    print("1. Performing division")
    x = 10 / 0   # Causes ZeroDivisionError

except ZeroDivisionError as e: # Catch the exception & assign it to variable 'e'
    print("2. Exception caught")
    print("Type of exception:", type(e))     # Class of exception
    # Output: <class 'ZeroDivisionError'>
    print("Arguments:", e.args)              # Tuple of arguments
    # Output: ('division by zero',)
    print("Message:", str(e))                # String representation
    # Output: division by zero

finally:
    print("3. Program continues")  # Always executes, use for cleanup
````

The script below shows how lists work in Python

````python
my_str = 'Hello World!' # A string assigned to variable my_str
print(str(my_str))  # Expected output: Hello World!
# str() converts to a string (although my_str is already a string)

print(my_str[0])  # Expected output: H
# Indexing starts from 0, so my_str[0] gives the first character.

print(my_str[1:5])  # Expected output: ello
# Characters from index 1 to index 4 (5 is excluded).
# Slice length = 5 - 1 = 4 characters.

print(my_str + ' From me')  # Expected output: Hello World! From me
# + joins two strings together.

print(my_str * 3)  # Expected output: Hello World!Hello World!Hello World!
# * repeats the string 3 times.

print(len(my_str))  # Expected output: 12
# len() returns the number of characters in the string.

print(my_str[-1])  # Expected output: !
# Negative indexing: -1 refers to the last character.

print(my_str.upper())  # Expected output: HELLO WORLD!
# .upper() converts all letters to uppercase.

print(my_str.lower())  # Expected output: hello world!
# .lower() converts all letters to lowercase.

print('World' in my_str)  # Expected output: True
# 'in' checks whether the substring exists inside the string.

print(my_str.replace('World', 'Python'))  # Expected output: Hello Python!
# replace(old, new) substitutes part of the string.
````

The script below shows how tuples work in Python

````python
pets = ['cat', 'dog', 'fish', 'rabbit', 'parrot', 'snake']
# A list of pet animals. Items are strings inside square brackets.

print(pets)
# Expected output: ['cat', 'dog', 'fish', 'rabbit', 'parrot', 'snake']
# Prints the whole list in its original order.

print(pets[0])
# Expected output: cat
# Indexing starts from 0, so pets[0] is the first item.

print(pets[1:5])
# Expected output: ['dog', 'fish', 'rabbit', 'parrot']
# Returns items from index 1 to index 4 (5 is excluded).

print(pets[-1])
# Expected output: snake
# -1 refers to the last item in the list.

print(len(pets))
# Expected output: 6
# len() returns how many items the list contains.

pets.append('hamster')
print(pets)
# Expected output: ['cat', 'dog', 'fish', 'rabbit', 'parrot', 'snake', 'hamster']
# append() adds an item to the end of the list.

pets.remove('fish')
print(pets)
# Expected output: ['cat', 'dog', 'rabbit', 'parrot', 'snake', 'hamster']
# remove() deletes the first matching item.

pets[2] = 'turtle'
print(pets)
# Expected output: ['cat', 'dog', 'turtle', 'rabbit', 'parrot', 'snake', 'hamster']
# Lists are mutable, so items can be modified by assigning a new value using an index.

print('dog' in pets)
# Expected output: True
# 'in' checks if an item is present in the list.

more_pets = ['frog', 'lizard']
print(pets + more_pets)
# Expected output: ['cat', 'dog', 'turtle', 'rabbit', 'parrot', 'snake', 'hamster', 'frog', 'lizard']
# + operator joins two lists together.
````

````python
# 1. Creating a tuple of animals
animals = ('cat', 'dog', 'rabbit', 'parrot')
print(animals) # Output: ('cat', 'dog', 'rabbit', 'parrot')
# Tuples are written inside parentheses () and items are separated by commas.

# 2. Accessing items with indexing
print(animals[0]) # Output: cat
# Indexing starts from 0, so animals[0] is the first item.

# 3. Accessing the last item using negative indexing
print(animals[-1]) # Output: parrot
# -1 refers to the last element in the tuple.

# 4. Slicing a tuple
print(animals[1:3]) # Output: ('dog', 'rabbit')
# Slice gives items from index 1 upto (but not including) index 3.

# 5. Checking the length of a tuple
print(len(animals)) # Output: 4
# len() gives the number of items in the tuple.

# 6. Tuples can store mixed data types
mixed = (10, 'hello', 3.5, True)
print(mixed) # Output: (10, 'hello', 3.5, True)
# A tuple can contain any valid Python object.

# 7. Tuples are immutable — trying to modify an item will cause an error
# animals[0] = 'lion' # Expected outcome: TypeError
# Explanation: Tuples are read-only; you cannot change their items.

# 8. A tuple with a single item must include a trailing comma
single = ('apple',)
print(single) # Output: ('apple',)
# Explanation: The comma is necessary to indicate that this is a tuple, not just a string in parentheses.
not_a_tuple = ('apple')
# Without the comma, Python treats it as a string inside parentheses.

# 9. Tuple packing (creating a tuple without parentheses)
student = 'John', 21, 'Physics'
print(student) # Output: ('John', 21, 'Physics')
# Python automatically packs the values into a tuple.

# 10. Tuple unpacking (assigning tuple items to variables)
name, age, subject = student
print(name, age, subject)  # Output: John 21 Physics
# Each variable receives the corresponding item from the tuple.
````

The script below shows use of Python dictionary

````python
# 1. Create dictionary student with key-value pairs inside {}.
student = {"name": "Alice", "age": 20, "course": "Maths"}
print(student) # {'name': 'Alice', 'age': 20, 'course': 'Maths'}

# 2. Accessing a value using its key inside square brackets
print(student["name"])  # Alice

# 3. Adding a new key-value pair. Dictionaries are mutable
student["year"] = 2026 # You can add new key/ value pairs anytime.
print(student) # {'name': 'Alice', 'age': 20, 'course': 'Maths', 'year': 2026}

# 4. Updating the value of an existing key.
student["name"] = "Bob"  # Assign a new value to update dictionary.
print(student)  # {'name': 'Bob', 'age': 21, 'course': 'Maths', 'year': 2026}

# 5. Removing a key-value pair
del student["course"]  # del removes a key/ value from dictionary.
print(student)  # {'name': 'Bob', 'age': 21, 'year': 2026}

# 6. Checking if a key exists using 'in' operator
print("age" in student)  # True

# 7. Getting the list of all keys using .keys()
print(student.keys()) # dict_keys(['name', 'age', 'year'])

# 8. Getting the list of all values using .values()
print(student.values()) # dict_values(['Bob', 21, 2026])

# 9. Getting key-value pairs together using .items()
print(student.items()) # dict_items([('name', 'Bob'), ('age', 21), ('year', 2026)])
# .items() returns each key and its value as a tuple.

# 10. Using .get() to safely access values
print("student get course:->", student.get("course")) # None
# .get() returns 'None' instead of causing an error if the key does not exist.
````

Sets Part 1:- Creating set from items, string, list. Also shows that a mutable item like a list cannot be a member of a set. Also shows how to (1) Find number of items (2) adding an intem (3) removing an item (4) Safely removing an item from a set.

````python
# 1. Creating a simple set of fruits
fruits = {"apple", "banana", "mango", "banana"}
print(fruits)  # {'apple', 'banana', 'mango'}
# Explanation: Duplicate 'banana' is removed because sets only keep unique elements.

# 2. Creating a set using a string
charSet = set("hello world")
print(charSet)  # {'d', 'e', 'h', 'l', 'o', 'r', 'w', ' '}
# Explanation: set() creates a set of unique characters from the string.

# 3. Creating a set from a list
numSet = set([1, 2, 2, 3, 4, 4, 5])
print(numSet)  # {1, 2, 3, 4, 5}
# Explanation: set() removes duplicate items automatically.

# 4. Error creating a set with mutable items
# The following line is commented out to avoid runtime error
# invalidSet = {1, 2, [3, 4]}  # Raises a TypeError
# Explanation: Lists are mutable and cannot be added to a set.

# 5. Checking the number of items in a set
print(len(fruits))  # 3
# Explanation: The set has 3 unique items.

# 6. Adding an item to a set
fruits.add("orange")
print(fruits)  # {'apple', 'banana', 'mango', 'orange'}
# Explanation: add() inserts new element (order not guaranteed).

# 7. Removing an item from a set using remove()
fruits.remove("banana")
print(fruits)  # {'apple', 'mango', 'orange'}
# Explanation: remove() deletes the specified item. Error if item not present.

# 8. Safely removing an item from a set using discard()
fruits.discard("papaya")  # papaya not in the set. But no error
print(fruits)  # {'apple', 'mango', 'orange'}
# Explanation: discard() does nothing if the item is missing.
````

Sets Part 2:- Shows (1) Checking for an item in a set (2) Union of two sets (3) Intersection of sets (4) Difference of two sets (5) Creating an empty set .

````python
# 1. Checking whether an item exists in a set
fruits = {"apple", "banana", "mango", "banana"}
print("mango" in fruits)  # True
# Explanation: 'in' checks membership in the set.

# 2. Union of two sets. Removes duplicates if present
A = {1, 2, 3}  # 3 is common in both A and B
B = {3, 4, 5}
print(A.union(B))  # {1, 2, 3, 4, 5}. 3 appears only once in result.
# Explanation: union() joins items from both sets.

# 3. Intersection of two sets
print(A.intersection(B))  # {3}
# Explanation: intersection() returns only common elements.

# 4. Difference of two sets
print(A.difference(B))  # {1, 2}
# Explanation: Items in A that are not in B.

# 5. Creating an empty set (IMPORTANT)
empty_set = set()
print(empty_set)  # set()
# Explanation: use set() to make an empty set.
````

Mutable vs Immutable Objects in Python. The following script shows that lists and dictionaries are mutable.


````python
# 1. Lists are mutable. Can be changed after they are created.
# Modifying list, does not create a new list; instead, it changes
# the existing list in place.
numbers = [1, 2, 3]
print("Before appending:", numbers, "id:", id(numbers))
# Output: Before appending: [1, 2, 3] id: 140353303123456
numbers.append(4)
# The list changed, but its id() stays the same.
print("After appending:", numbers, "id:", id(numbers))
# Output: After appending: [1, 2, 3, 4] id: 140353303123456

# 2. Dictionaries are mutable. They can be changed after they are created.
d = {"a": 1, "b": 2}
print(id(d))  # 140733544647664
d["c"] = 3  # Modifying dictionary by adding new key-value pair
print(id(d))  # 140733544647664 (same as previous id)
# Explanation: Dictionary was modified in-place (same object).
````

The following script shows that Python integers and strings are immutable.

````python
# 1. Immutable example using integers
x = 10
print(id(x))  # 140733544647384
x = x + 5
print(id(x))  # 140733544647456 (different from previous id)
# Explanation: NEW int object created because ints are immutable

# 2.Immutable example using strings
s = "hello"
print(id(s))  # 140733544647520
s = s + " world"  # Modifying string creates a NEW string object
print(id(s))  # 140733544647584 (different from previous id)
# Explanation: Strings cannot change; Python creates NEW string object.

````

**Implicit Type Casting** (Python does it automatically):-
The following code explains implicit type casting


````python
# 1. Addition of int and float results in float (type promotion)
x = 4        # int
y = 3.5      # float
result = x + y  # Python promotes x to float and performs float addition
print(result)  # 7.5
# Explanation: Python converts 4 to 4.0 and performs float addition.

# 2. Multiplication of int and float results in float (type promotion)
a = 2.0
b = 5
print(a * b)  # 10.0
# Can multiply float and int, result is float.
# Explanation: Python promotes 5 to 5.0 and performs float multiplication.

# 3. Division operator always returns a float
print(5 / 2)  # 2.5
# Explanation: Even if both are ints, Python promotes the result to float.
````

Examples: Type Conversion to bool
Below are simple examples showing how Python converts different values into boolean values.

````python
print(bool(5))        # True  -> non-zero integer is True
print(bool(-6))       # True  -> negative non-zero integer is True
print(bool(0))        # False -> zero is False

print(bool("hello"))  # True  -> non-empty string is True
print(bool(""))       # False -> empty string is False
print(bool(" "))      # True  -> string with space is NOT empty

print(bool(0.0))      # False -> float zero is False
print(bool(3.14))     # True  -> any non-zero float is True

print(bool([]))       # False -> empty list is False
print(bool([1, 2, 3]))# True  -> non-empty list is True

print(bool({}))       # False -> empty dictionary is False
print(bool({"a": 1})) # True  -> non-empty dictionary is True

print(bool(None))     # False -> None is always False
````

````python
# Ways to read Documentation. Use math module as an example.
# 1. Using help() on math module
import math
help(math)
# Displays full documentation of math module
# Shows functions, attributes, and explanations

# 2. Using dir() to list contents of math module
print(dir(math))
# Lists ALL names defined inside math module
# Includes functions (sqrt, sin, cos) and attributes (pi, e)

# 3. Reading docstring of a function sqrt() of math module
print(math.sqrt.__doc__)
# Displays the documentation string of sqrt()
# Shows short description of what the function does

# 4. Using help() on factorial() function of math module
help(math.factorial)
# Shows detailed help for factorial(). Explains parameter
# requirements (For factorial, integer must be >= 0)

# 7. Using docstring of math module
print(math.__doc__)
# Shows module-level documentation string
````

````python
# Script 2: Reading Documentation of Third-Party Python Modules

# This script demonstrates how to:
# 1. Check whether a third-party module is installed
# 2. Read its documentation safely
# 3. Use help(), dir(), and docstrings
# 4. Use pydoc programmatically
# We use try / except so that the program does NOT crash
# if a third-party module is not installed.

# A. Reading documentation for NumPy (third-party module)

try:
    import numpy as np

    print("\nNumPy module found!")  # Will print if NumPy is imported successfully
    print("Exploring NumPy documentation...\n")

    # dir() lists all top-level names in the module
    print("Listing some NumPy attributes and functions:")
    print(dir(np))     # Shows array, zeros, ones, random, etc.

    print("\nHelp on numpy.array:")
    help(np.array)     # Detailed documentation for array()

    print("\nDocstring of numpy.array:")
    print(np.array.__doc__)   # Short description of array constructor

except ModuleNotFoundError:
    print("NumPy is NOT installed.")
    print("Install it using: pip install numpy")

# C. Using pydoc programmatically (built-in module)
import math
import pydoc

# pydoc.render_doc() generates formatted documentation as text
math_doc = pydoc.render_doc(math)  # Get documentation for math module as a string

# Print only the first 500 characters to keep output readable
print(math_doc[:500])
# pydoc is useful when:
# - You want offline documentation
# - You want to save docs to a file
# - You want clean text output

# D. Opening documentation in a web browser (Optional)
# pydoc can also start a local documentation web server.
# This is useful for browsing documentation in a browser.

# Uncomment the line below if you want to open documentation
# in your default web browser.
pydoc.browse()
# Message like Server ready at http://localhost:59635/
# This will open a local web page with documentation for all installed modules.
# You can search for any module, including third-party ones, in that web interface.
````

Function Overloading in Python
Does Python support Function Overloading? Technically, no.

```python
# --- First Definition of the function ---
def greet():
    print("Hello! This is the first version of the function.")

# --- Second Definition of the same function name ---
# This version takes one argument with a default value.
# In Python, this second block completely replaces the first one.
def greet(name="Student"):
    print(f"Hi {name}! This is the second version of the function.")

# --- Testing the function ---
# Even if we call it with no arguments (which the 1st version expected),
# Python only 'knows' the 2nd version exists.
greet()  # Output: Hi Student! This is the second version of the function.
greet("Alice")  # Output: Hi Alice! This is the second version of the function.
```

The following script creates a simple Pet class and an instance of the class named my_dog

```python

class Pet:
    # Statement 1: Defining a Class Attribute (shared by all)
    species = "Canine"

    # Statement 2: Defining an Instance Method (a behavior)
    def bark(self):
        print("Woof!")

# --- Using the Class ---

# 1. Accessing data directly from the Class (The Blueprint)
print(Pet.species)    # Output: Canine

# 2. Creating an Instance (The Object)
my_dog = Pet()

# 3. Accessing data and behavior from the Instance
print(my_dog.species) # Output: Canine
my_dog.bark()         # Output: Woof!
```

The following script shows the usage of __init__() in a class. Using __init__() personalizing our Pets to create an instance/ object with attributes unique to the instance/ object.

```python
class Pet:
    # The Constructor: Takes 4 parameters (including self)
    def __init__(self, name, breed, age):
        # Assigning the parameters to 'self' (the object)
        self.name = name   # Unique name for this instance
        self.breed = breed # Unique breed for this instance
        self.age = age     # Unique age for this instance

    def display_info(self):
        print(f"Pet Name: {self.name}, Breed: {self.breed}, Age: {self.age}")

# --- Creating unique objects ---
# Note: We pass 3 arguments but __init__ sees 4. Python adds 'self'
# automatically. Each object has its own data (name, breed, age)
# stored in 'self', which is unique to that instance.
p1 = Pet("Tiger", "Labrador", 3)  # The first pet object
p2 = Pet("Kittie", "Siamese Cat", 2)  # The second pet object

# Printing attributes directly
print(f"p1 is {p1.name}, a {p1.breed}.")  # Output: p1 is Tiger, a Labrador
print(f"p2 is {p2.name}, a {p2.breed}.")  # Output: p2 is Kittie, a Siamese Cat

# Calling a method that uses 'self'
p1.display_info()  # Output: Pet Name: Tiger, Breed: Labrador, Age: 3
p2.display_info()  # Output: Pet Name: Kittie, Breed: Siamese Cat, Age: 2
```

The following script shows the use of constructor with default parameters:

```python
# This code demonstrates the use of default values in a class constructor

class Pet:
    # No default value  for name. But defaults for species and age given.
    def __init__(self, name, species="Dog", age=1):
        self.name = name
        self.species = species
        self.age = age

# 1. Using both Defaults: Only providing the name Tommy.
# species defaults to "Dog", age defaults to 1.
p1 = Pet("Tommy")
print(f"{p1.name} is a {p1.species} (Age: {p1.age})")
# Output: Tommy is a Dog (Age: 1)

# 2. Overriding 1 Default namely age: Provide name and species.
# species overrides the default "Dog", but age still defaults to 1.
p2 = Pet("Luna", "Cat")  # Creating an instance with name and species. Age takes the default value.
print(f"{p2.name} is a {p2.species} (Age: {p2.age})")
# Output: Luna is a Cat (Age: 1)

# 3. Full Customization: Provide all three arguments. No default used
p3 = Pet("Fluffy", "Rabbit", 5)  # Creating instance with name, species & age
print(f"{p3.name} is a {p3.species} (Age: {p3.age})")
# Output: Fluffy is a Rabbit (Age: 5)
```

Class with Methods: Defining the Behavior of a Class

```python
class Pet:
    # Constructor with a default value for species and an additional attribute
    # mood which not in the constructor parameters. So when you create a Pet
    # object, you only provide name and optionally species. You cannot set
    # mood at creation, it always starts as "Happy". But you can change it
    # later using the setter method..
    def __init__(self, name, species="Dog"):
        self.name = name  # This attribute is set from the constructor parameter
        self.species = species  # This has default value of "Dog" if not given
        # This attribute is NOT in the constructor parameters.
        # Every pet starts with a default mood of 'Happy'.
        self.mood = "Happy"  # This is an attribute that is initialized in the
        # constructor but not passed as a parameter.

    # Method to update the mood (Setter)
    def set_mood(self, new_mood):
        # We use self.mood to change the attribute for this specific pet
        self.mood = new_mood

    # Method to retrieve the mood (Getter)
    def get_mood(self):
        # Returns the current value of the attribute
        return self.mood

# --- Testing the Methods ---

# 1. Create a pet object
my_pet = Pet("Rocky")  # We only provide the name. Species defaults to "Dog".
# Mood defaults to "Happy".
print(f"{my_pet.name} is a {my_pet.species}.")  # Output: Rocky is a Dog.

# 2. Check the default mood using the getter method
print(f"Default mood: {my_pet.get_mood()}") # Output: Default mood: Happy

# 3. Change the mood using the setter method
my_pet.set_mood("Sleepy")  # This changes mood attribute of my_pet to "Sleepy"

# 4. Check the mood again
print(f"Updated mood: {my_pet.get_mood()}")  # Output: Updated mood: Sleepy
```

The following script shows use of (1) Class variable count (2) Use of decorator @staticmethod.

```python
class Pet:
    count = 0

    def __init__(self, name, species="Dog"):
        Pet.count += 1 # A shorthand for Pet.count = Pet.count + 1
        self.name = name
        self.species = species
        print(f"Created {self.name}. Total pets now: {Pet.count}")

    # 1. The Modern Decorator tells Python this is a Static Method
    @staticmethod
    def display_total():
        # This belongs to the Class, no 'self' needed
        print(f"--- Global Pet Count: {Pet.count} ---")

# --- Using the Class ---
p1 = Pet("Tommy")  # Create first pet object. Increments count to 1.
p2 = Pet("Luna", "Cat")  # Create second pet object. Increments count to 2.

# 2. Calling the static method. No object needed, we call it on class itself.
Pet.display_total()  # Output: --- Global Pet Count: 2 ---
```

The Script: Decorating a Class

```python
# --- THE CLASS DECORATOR ---
def add_pet_status(cls):
    # A decorator that 'injects' extra info into a class blueprint

    # 1. Add simple attributes
    cls.location = "Delhi"
    cls.is_vaccinated = True

    # 2. Add a simple method which uses the injected attribute
    def display_status(self):
        if  self.is_vaccinated:
            status = "Protected"
        else:
            status = "Not Protected"

        return f"{self.name} is at {self.location}. Status: {status}"

    # Attach method to the class. Inject it into class blueprint. If you
    # don't do this, the method won't be available to instances of the class
    cls.display_status = display_status

    # 3. Return the class (The Blueprint is now upgraded)
    return cls

# --- APPLYING THE DECORATOR ---
@add_pet_status
class Dog:
    def __init__(self, name):
        self.name = name

# --- TESTING ---
my_dog = Dog("Tiger")

print(f"Location: {my_dog.location}")      # Directly access injected attribute
# Output: Location: Delhi
print(f"Summary:  {my_dog.display_status()}") # Call the injected method
# Output: Summary:  Tiger is at Delhi. Status: Protected
```

The following script shows the use of name mangling

```python
# This code demonstrates the concept of name mangling in Python classes.
class Pet:

    name = "Cat"          # normal variable
    __type = "Animal"     # mangled variable

p = Pet()
print("Normal variable:", p.name)  # Output: Normal variable: Cat
print("Mangled variable:", p._Pet__type)  # Output: Mangled variable: Animal
```

Following script shows use of __doc__ ,  __bases__ and __str__()

```python
class Pet:
    """<<Pet class docstring>>"""

    def __init__(self, name):
        self.name = name

print("Class docstring ->", Pet.__doc__)
# Output: Class docstring -> <<Pet class docstring>>

print("Base class of Pet ->", Pet.__bases__)
# Output: Base class of Pet -> (<class 'object'>,)

# Create an object
p = Pet("Tom")

print("Object representation using print() ->", p)
# print() internally calls the __str__() method

print("Calling __str__() explicitly ->", p.__str__())
# Output will be something like: <__main__.Pet object at 0x...>
```

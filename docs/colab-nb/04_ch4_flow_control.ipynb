The general form of the if statement in Python looks like this:

```python

if condition_1:# condition_1 must be a Bool expression and eval to True or False
    statement_block_1
elif condition_2:	# condition_2 also must be a bool expression
    statement_block_2
else:
    statement_block_3
    
```

The following script takes an input, converts it into an integer. It then uses if--- elif--- else blocks to check if it is positive, negative or 0

````python
# --- STEP 1: Get User Input ---
user_entry = input("Enter a number: ")

# --- STEP 2: Convert to Integer ---
# We convert the input string directly to a number using int().
# Note: Will raise an error if input not a valid integer,
# but we are skipping error handling for simplicity.
num = int(user_entry)

# --- STEP 3: The Decision Structure (Selection Flow) ---
if num > 0:
    # Runs if num is greater than zero
    print(f"Success: {num} is a positive number.")

elif num < 0:
    # Runs if num is less than zero
    print(f"Success: {num} is a negative number.")

else:
    # Runs only if the number is exactly zero
    print("Success: The number is zero.")

# --- STEP 4: End of Script ---
print("Program execution finished.")
````

The following script checks whether a number is even or odd.

````python
# --- STEP 1: Get User Input ---
# We use input() to get a string from the user and int() to convert it
user_entry = input("Enter a whole number to check: ")
num = int(user_entry)

# --- STEP 2: The Two-Way Decision (Selection Flow) ---
# We use the Modulo Operator (%) to find the remainder.
# If a number divided by 2 has a remainder of 0, it is EVEN.

if num % 2 == 0:
    # This block executes if the boolean expression evaluates to True
    print(f"The number {num} is Even.")
else:
    # This block executes if the boolean expression evaluates to False
    print(f"The number {num} is Odd.")

# --- STEP 3: Sequential Flow Resumes ---
print("Flow Control test complete.")

````

#### Omitting the else Clause (The "One-Way" Decision)

````python
# --- STEP 1: Set the Temperature (Data) ---
temp = 30  # Try changing this to 20 to see the skip in action

# --- STEP 2: The One-Way if Statement ---
if temp > 25:
    # This block ONLY runs if temp is greater than 25
    print(f"Current temperature is {temp}°C.")
    print("Action: Switching the Air Conditioner ON.")

# --- STEP 3: Rest of the Script ---
# This line is NOT indented.
# It runs regardless of whether the 'if' was True or False.
print("Monitoring system is active...")
````

#### Nesting
Nesting occurs when you place an if...elif...else statement inside the block of another if...elif...else statement.


The following script shows use of nested if---else blocks:-


````python
# --- STEP 1: Get User Input ---
# We convert the input to an integer immediately
my_age = int(input("Enter Age: "))

# --- STEP 2: The Nested Logic ---
# Each 'else' block contains a new 'if' statement,
# pushing the logic deeper into the nest.
if my_age >= 18:
    my_comment = 'You can vote'
else:
    # First level of nesting: Happens if age is < 18
    if my_age >= 13:
        my_comment = 'You are a teenager'
    else:
        # Second level of nesting: Happens if age is < 13
        if my_age >= 10:
            my_comment = 'You are in middle school'
        else:
            # Third level of nesting: Happens if age is < 10
            if my_age >= 6:
                my_comment = 'You are in primary school'
            else:
                # The deepest part of the nest
                my_comment = 'You are too small to learn Python'

# --- STEP 3: Sequential Flow Resumes ---
# This line is unindented, so it always executes at the end.
print(f"At age: {my_age} -> {my_comment}")
````

#### How to Avoid Nested if Statements (Good Programming Technique)

By combining multiple conditions into a single statement, you reduce the indentation level and make the logic much clearer at a glance.

**Example:** **The "Teenager" Logic.** Instead of checking the age twice in a "staircase" of code, we use a single, readable line. So instead of testing the `if` condition twice, we have combined them into a single `if` condition using the logical operator and. We can also use Python's Comparison Chaining. The following script shows this.


````python
# --- THE REFACTORED APPROACH ---
# We assume 'age' is already defined as an integer.
age = 15

# Option 1: Using 'and' to Combine Conditions combine lower bound (>12)
# and upper bound (<20).Much cleaner than nesting one 'if' inside another.
if age > 12 and age < 20:
    # This block runs only if BOTH conditions are met.
    print(f"Age {age}: Status is 'Teenager'.")

# Option 2: Using Python's Comparison Chaining. Even cleaner
if 12 < age < 20:
    print("This is also valid and very readable!")
````

We can use this technique to avoid the nested if-else loop in the example of two numbers and testing them for even or odd as follows:-

````python
# --- STEP 1: Get User Input ---
# We convert the inputs to integers to perform the modulo calculation.
num1 = int(input('Enter the first number: '))
num2 = int(input('Enter the second number: '))

# --- STEP 2: Multi-way Decision using 'and' ---
# Each 'elif' represents a specific combination of conditions.

if (num1 % 2 == 0) and (num2 % 2 == 0):
    # Runs only if BOTH numbers have a remainder of 0
    print('Both numbers are Even.')

elif (num1 % 2 != 0) and (num2 % 2 == 0):
    # Runs if num1 is Odd AND num2 is Even
    print('First number is Odd, second number is Even.')

elif (num1 % 2 == 0) and (num2 % 2 != 0):
    # Runs if num1 is Even AND num2 is Odd
    print('First number is Even, second number is Odd.')

else:
    # If none of the above are True, both must be Odd.
    # Note: No boolean expression is needed for 'else'.
    print('Both numbers are Odd.')

# --- STEP 3: Sequential Flow Resumes ---
print("Comparison complete.")
````

#### The `while` Loop (Iterative Flow)
The `while` loop is used to repeat a block of code as long as a specific condition remains `True`.

Unlike an `if` statement that runs once, the while loop checks the condition, executes the body, and then "loops" back to check the condition again.

Syntax of while loop with the optional else is as below:-

```python

while test_condition:   # Must evaluate to True or False
    # Loop body: Runs repeatedly while condition is True
    statements
else:                   
    # Optional: Runs once when the condition becomes False
    statements

```


An example of use of while loop is given below (It prints the first five numbers and their squares in two columns seperated by tabs):-

````python
# --- STEP 1: Initialize the Counter ---
# We start x at 1. This variable controls how many times the loop repeats.
x = 1

# --- STEP 2: Print the Table Header ---
# \t is an 'escape character' that acts like the Tab key on your keyboard.
print('Number\tSquare')
print('------\t------')

# --- STEP 3: The while Loop ---
# The condition (x <= 5) is checked BEFORE every iteration.
while x <= 5:

    # Calculate the square (x to the power of 2)
    square = x ** 2

    # Print the result using an f-string (formatted string).
    # This automatically converts numbers to strings for us!
    print(f"{x}\t{square}")

    # --- STEP 4: Update the Counter ---
    # We add 1 to x so the loop eventually reaches 6 and stops.
    # If we forgot this line, the loop would run forever (Infinite Loop)!
    x = x + 1

# --- STEP 5: Loop Exit ---
# This line is unindented, so it only runs once the loop is finished.
print('Done!')
````

#### Another example:-
The script below is a good way to demonstrate how a while loop can interact with string slicing to process data step-by-step. The script successively strips the first 2 characters of a string and print result. Note when string `s1` becomes empty, the while loop will terminate


````python
# --- STEP 1: Define the initial string ---
s1 = '112233'

# --- STEP 2: The Loop Condition ---
# In Python, an empty string '' is considered False.
# A string with characters is considered True.
# Therefore, 'while s1:' means 'Keep going until the string is empty'.
while s1:
    # Print the current state of the string
    print(s1)

    # --- STEP 3: String Slicing (The Update) ---
    # s1[2:] creates a NEW string starting from index 2 to the end.
    # This effectively "chops off" the first two characters.
    s1 = s1[2:]

# --- STEP 4: Loop Exit ---
print("The string is now empty!")
# Output:
# 112233
# 2233
# 33
# The string is now empty!
````

#### Control Flow Tools: break, continue, and pass

The following script shows the use of (1) break (2) continue and (3) pass

````python
x = 0
print("Starting the loop...")

while x < 10:
    x = x + 1

    # 1. BREAK STATEMENT Exits loop immediately even if condition is still True.
    if x == 8:
        print(f"--- Found {x}: Breaking (Exiting loop entirely) ---")
        break

    # 2. CONTINUE STATEMENT Skips rest of current iteration. Moves to next
    if x == 5:
        print(f"--- Found {x}: Continuing (Skipping rest of this iteration) ---")
        continue

    # 3. THE PASS STATEMENT Does nothing, acts as a placeholder.
    if x == 3:
        # We know 3 is special, but we haven't decided what to do yet.
        # 'pass' keeps the code from crashing.
        pass
        print(f"--- Found {x}: Pass was used, but execution continues below ---")

    # This print statement is skipped when 'continue' or 'break' is triggered
    print(f"Current number is: {x}")

print("Loop is finished!")
````

#### The continue Statement:
While break is an emergency exit that kills the loop entirely, continue is more like a skip button. It tells Python: "Stop what you are doing in this specific round, ignore the code below me, and jump back to the top for the next item."

The following script shows the use of continue statement

````python
# --- THE CONTINUE DEMONSTRATION ---

print("Starting the countdown (skipping unlucky 7)...")

for num in range(1, 11):
    if num == 7:
        print("--- SKIPPING 7: Jumping to the next number! ---")
        continue  # Everything below this line is ignored for '7'

    # This line only runs if 'continue' was NOT triggered
    print(f"Processing number: {num}")

print("Done!")
````

#### The match Statement (Python 3.10+)

It is Python’s version of the "Switch" statement found in C++ or Java. Introduced in Python 3.10,  is a cleaner alternative to a long chain of if...elif...else statements.
The following script shows the use of match statements. The script asks user to press only a single key.
* (1) If the key pressed is an alphabet, it matches case for alphabet
* (2) If the key pressed is a number it matches the case for a number
* (3) If any other key or multiple keys are pressed, it goes for the "Wildcard" case.


````python
# --- ADVANCED MATCH-CASE: PATTERN GUARDS ---

key = input("Press a single key: ")

match key:
    # 1. Case for an Alphabet (a-z or A-Z)
    # We use 'if' inside the case to add a condition (a "Guard")
    case char if char.isalpha() and len(char) == 1:
        print(f"Success: '{char}' is an Alphabet.")

    # 2. Case for a Number (0-9)
    case num if num.isdigit() and len(num) == 1:
        print(f"Success: '{num}' is a Number.")

    # 3. Wildcard Case for any other key (Symbols, Space, etc.)
    case _:
        print("Note: You pressed a special character or multiple keys.")

print("Key analysis complete.")
````

#### Example 1: Using while to mimic if



````python
# --- Script: Check Even/Odd using while...else ---
user_input = input('Give a number -> ')
num = int(user_input)

while num % 2 == 0:
    # This block runs if the condition is True (Even)
    print(f"{num} is even")

    # IMPORTANT: We must 'break' here.
    # If we don't, 'num' stays even forever, and we get an Infinite Loop!
    break

else:
    # This block runs if the condition was False (Odd)
    # OR if the loop finished without hitting a 'break'
    print(f"{num} is odd")
````

#### Example 2: A Practical Use Case (Searching)
The else clause is most helpful when searching for something.

The following script shows this:-

````python
# --- Script: Search for the letter 'z' ---
word = "Python"  # Change this to test with different words,"Zebra"
index = 0

while index < len(word):
    if word[index] == 'z':
        print("Found a 'z'!")  # This will print if 'z' is found
        break # Exit the loop and SKIP the else block
    index += 1
else:
    # This only runs because the loop finished without finding 'z'
    print("Search complete: No 'z' was found in this word.")

print("End of program.")
````

#### Simulating a do...until Loop in Python
Example: Data Validation (The "Input Trap"). A common use for this pattern is Input Validation. This "traps" the user in a loop until they provide the correct type of data.

````python
# --- SIMULATING DO...UNTIL ---
# We want to force the user to enter a valid number.

while True:
    # 1. Ask for input (This always runs at least once)
    my_age = input('Give your age in years -> ')

    # 2. Check the condition (Is the input composed only of digits?)
    if my_age.isdigit():
        # Success! Print the result and exit the loop.
        print(f'Success: You are {my_age} years old.')
        break

    # 3. This runs only if the 'if' condition was False (break didn't happen)
    print('Error: You did not enter a valid integer. Please try again.\n')

print('Program continues...')
````

#### Looping over Python collections/ sequences

The following script shows how the for loop is used to loop over a Python string.
(Similarly one can loop over other objects like lists, tuples and dictionaries)

````python
# --- STEP 1: Define a collection ---
word = "Python"

print(f"Let's spell '{word}' one letter at a time:")
print("--------------------------------")

# --- STEP 2: The for Loop ---
# 'letter' is our loop variable.
# It will take the value of each character in 'word' one by one.
for letter in word:
    # In the 1st round, letter = 'P'
    # In the 2nd round, letter = 'y', and so on.
    print(f"Current character: {letter}")

else:
    # This runs only after the loop has visited EVERY letter.
    print("--------------------------------")
    print("Loop complete: All letters processed!")

# --- STEP 3: Outside the loop ---
print("Program continues...")
````

#### Example script:- Finding Matching Characters

````python
# --- STEP 1: Define the two strings to compare ---
str1 = "hello"
str2 = "world"

print(f"Comparing: '{str1}' & '{str2}'")
print("----------------------------")

# --- STEP 2: The Outer Loop ---
# This picks one character (ch1) from the first string at a time.
for ch1 in str1:

    # --- STEP 3: The Inner Loop ---
    # For the current ch1, we look at every character (ch2) in str2.
    for ch2 in str2:

        # --- STEP 4: The Comparison ---
        # We check if the character from the first word matches the second.
        if ch1 == ch2:
            print(f"Match found: '{ch1}' is in both words.")

# --- NOTE TO STUDENTS ---
# Notice that 'l' is printed multiple times.
# This happens because 'l' appears twice in 'hello'
# and once in 'world'. The loop finds EVERY possible match!
````

#### Dead Code (Unreachable Code)
The following script shows example of dead code:-

````python
# Example: Identifying Dead Code

# --- Type 1: Post-Termination Code ---
while True:
    print("This line runs!")
    break
    print("DEAD CODE: This will never print because of the break above.")

# --- Type 2: Impossible Conditions ---
x = 10
if x < 5 and x > 15:
    print("DEAD CODE: A number cannot be less than 5 AND greater than 15.")

# --- Type 3: Redundant Logic ---
# Imagine a student's score is 95
score = 95
# 1. This condition is checked first.
# Since 95 is greater than 50, this block executes.
if score > 50:
    print("You passed!")

# 2. DEAD CODE / REDUNDANT BLOCK:
# Even though 95 is also greater than 90, Python will NEVER look at this elif.
# Once the first 'if' is True, Python skips all related 'elif' and 'else' blocks.
elif score > 90:
    print("Excellent! You got an A!")

# 3. This else is also skipped because the first 'if' was True.
else:
    print("Better luck next time.")

print("Program finished.")
````

The following scripts shows the different ways in which the range() function is used:-


````python
# ==========================================================
# Demonstrating various usages of the range() function
# ==========================================================

# 1. Using range() with a SINGLE argument (stop)
# Starts at 0 by default. Ends at (stop - 1).
print("Single argument (0 to 4):")
print(list(range(5)))         # Output: [0, 1, 2, 3, 4]

# 2. Using range() with START and STOP arguments
# Starts at 2. Ends at (6 - 1 = 5).
print("\nStart and Stop (2 to 5):")
print(list(range(2, 6)))      # Output: [2, 3, 4, 5]

# 3. Using range() with START, STOP, and STEP arguments
# Starts at 0. Increments by 3. Stops before 12.
print("\nWith positive step (0 to 9):")
print(list(range(0, 12, 3)))  # Output: [0, 3, 6, 9]

# 4. Using range() with NEGATIVE STEP (Counting Backwards)
# Starts at 0. Decrements by 1. Stops before -5.
print("\nNegative step (0 to -4):")
print(list(range(0, -5, -1))) # Output: [0, -1, -2, -3, -4]

# 5. Negative START and STOP with NEGATIVE STEP
# Starts at -2. Decrements by 2. Stops before -12.
print("\nNegative start/stop/step (-2 to -10):")
print(list(range(-2, -12, -2))) # Output: [-2, -4, -6, -8, -10]

# ----------------------------------------------------------
# LOGIC TRAPS: Why some ranges return an empty list []
# ----------------------------------------------------------

# 6. Impossible Range: Negative start/stop with POSITIVE step
# Logic: You cannot count UP from -2 to reach -12.
print("\nImpossible (Start > Stop with positive step):")
print(list(range(-2, -12, 2)))  # Output: []

# 7. Impossible Range: Positive start/stop with NEGATIVE step
# Logic: You cannot count DOWN from 2 to reach 8.
print("\nImpossible (Start < Stop with negative step):")
print(list(range(2, 8, -1)))    # Output: []

print("\n--- End of Demonstration ---")
````

#### Using `range()` function in loop

````python
# --- 1. Looping over a String using Indexes ---
# We use len() to find the total characters, and range() to get the index numbers.
s = "Hello!"
print("String iteration by index:")
for i in range(len(s)):
    print(f"Index {i} contains character: {s[i]}")

# --- 2. Looping over a List ---
my_list = [10, 20, 30, 40]
print("\nList iteration by index:")
for i in range(len(my_list)):
    print(f"Item at index {i} is {my_list[i]}")

# --- 3. Custom Steps (Counting by 2s) ---
print("\nEven numbers from 0 to 8:")
for i in range(0, 10, 2):  # Starts at 0, stops before 10, jumps by 2
    print(i, end=" ")

# --- 4. Reverse Order (Countdown) ---
print("\n\nCountdown from 10 down to 1:")
for i in range(10, 0, -1): # Start 10, Stop 0 (exclusive), Step -1
    print(i, end=" ")

# --- 5. Negative Start and Stop ---
print("\n\nNegative range (-1 down to -5):")
for i in range(-1, -6, -1):
    print(i, end=" ")

# --- 6. Large Steps ---
print("\n\nStep larger than the range:")
for i in range(1, 10, 15):
    print(i) # Only prints 1, then stops because 1 + 15 > 10

# NO OUTPUT:- Empty Ranges. Dont cause errors. They just don't run the loop.

# A. Equal Start and Stop: Nothing to loop over.
print("\nEqual start and stop (3, 3):")
for i in range(3, 3):
    print("This will not print")

# B. Impossible Direction: You can't go "down" from 1 to 5 with a negative step.
print("Impossible direction (1 to 5 with -1):")
for i in range(1, 5, -1):
    print("This will not print")

# C. Range(0): This means "loop zero times."
print("Range(0):")
for i in range(0):
    print("This will not print")

# WHAT CAUSES A "VALUE ERROR"? a STEP of 0.
# Because you cannot "move" by zero, Python stops the script.

# for i in range(1, 10, 0):  <-- THIS WOULD CAUSE AN ERROR
#    print(i)

print("\n--- End of Exercises ---")

````

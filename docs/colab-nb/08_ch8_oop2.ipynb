The following script has a class called Parent. Another class Child is derived from Parent. Then instances of both Parent (Named p1) and Child (Named c1) are created. Then type() function is show the "type" of classes p1 and c1 are. Further, the isinstance() function is used to see whether c1 and p1 belong to Parent or Child or not. The output of isinstance() shows that c1 is instance of both Parent and Child class. But p1 is only instance of Parent class and is not an instance of Child class.

```python
class Parent:
    pass

class Child(Parent):  # Child inherits from Parent.
    # Child is a subclass of Parent and Parent is a superclass of Child
    pass

p1 = Parent()  # Create an instance of Parent
c1 = Child()  # Create an instance of Child

# Let's analyze the types and inheritance relationships
# type() will show us the class of the instance
print("type of p1:", type(p1))  # <class '__main__.Parent'>
print("type of c1:", type(c1))  # <class '__main__.Child'>

# isinstance() will check if an instance belongs to a class or its subclasses
print(f"Is p1 an instance of Parent? {isinstance(p1, Parent)}")  # True
print(f"Is c1 an instance of Child? {isinstance(c1, Child)}")  # True
print(f"Is c1 an instance of Parent? {isinstance(c1, Parent)}")
# True, because Child inherits from Parent
print(f"Is p1 an instance of Child? {isinstance(p1, Child)}")
# False, because Parent does not inherit from Child
```

The following script has a class Animal which implicitly inherits from object class. Further there is a class Dog which inherits from Animal class. The mro for object, Animal and Dog classes are shown in this script:-

```python
print(object.__mro__)  # Shows the mro for the base 'object' class
# Output: (<class 'object'>,)
# So object → object (itself)

class Animal:
    def speak(self):
        print("Animal speaks")

print(Animal.__mro__)  # Shows the method resolution order for the Animal class
# Output: (<class '__main__.Animal'>, <class 'object'>)
# So Animal → object

class Dog(Animal):
    pass

print(Dog.__mro__)  # Shows the method resolution order for the Dog class
# Output: (<class '__main__.Dog'>, <class '__main__.Animal'>, <class 'object'>)
# So Dog → Animal → object
```

The following script shows use of super() in calling constructor and calling a method of the parent class.

```python
# Demonstration of super() with multi-level inheritance (MRO)

class Pet:
    def __init__(self, name):
        self.name = name  # Instance variable to store name of pet
        print("Pet constructor called")

    def walk(self):  # Parent method
        print(f"{self.name} is walking (Pet method)")

class Dog(Pet):  # Dog inherits from Pet
    def __init__(self, name):  # Overrides the constructor of Pet
        print("Dog constructor called")
        super().__init__(name)  # Calls Pet constructor

    def walk(self):  # Overriding + super() to call parent method
        print(f"{self.name} is walking (Dog method)")
        super().walk()          # Calls next in MRO → Pet

# Create object of Dog
d = Dog("Tommy")  # Constructor chaining: Dog → Pet
# Expected Output: on object creation:
# Dog constructor called
# Pet constructor called
print("Calling walk():")
d.walk()  # Expected Output:
# Tommy is walking (Dog method)
# Tommy is walking (Pet method)

# NEW CLASS (Derived from Dog)
class Puppy(Dog):  # Puppy inherits from Dog
    def __init__(self, name):  # Overrides the constructor of Dog
        print("Puppy constructor called")
        super().__init__(name)  # Calls Dog → then Pet

    def walk(self):  # Overriding + super() to call parent method
        print(f"{self.name} is walking (Puppy method)")
        super().walk()          # Calls Dog → then Pet

# Create object of Puppy
p = Puppy("Tommy")  # Constructor chaining: Puppy → Dog → Pet
# Expected Output: on object creation:
# Puppy constructor called
# Dog constructor called
# Pet constructor called

print("Calling walk():")
p.walk()  # Expected Output:
# Tommy is walking (Puppy method)
# Tommy is walking (Dog method)
# Tommy is walking (Pet method)

print("MRO of Puppy:")  # MRO shows order in which classes are looked up when
# searching for a method. MRO for Puppy is Puppy → Dog → Pet → object.
print(Puppy.__mro__)  # Output:(<class '__main__.Puppy'>,
# <class '__main__.Dog'>, <class '__main__.Pet'>, <class 'object'>)
```

The following script shows the 4 types of IS-A inheritence:

```python
class Pet:
    def __init__(self, name):  # Constructor
        self.name = name

    def speak(self):  # Method to be overridden
        print(f"{self.name} makes a generic pet sound.")

    def eat(self):  # Method to demonstrate code reuse
        print(f"{self.name} is eating.")

# --- TYPE 1: Pass-through (Verbatim) ---
# Logic: "I am exactly like a Pet."
class BasicPet(Pet):  # No new attributes/ methods. Inherits everything as-is
    pass

# --- TYPE 2: Additive (Extension) ---
# Logic: "I am a Pet, but I also have a breed and can fetch."
class Retriever(Pet):  # Inherits from Pet and adds new behavior
    def fetch(self):
        print(f"{self.name} is fetching the ball!")

# --- TYPE 3: Replacement (Overriding) ---
# Logic: "I am a Pet, but I don't make generic sounds; I ONLY purr."
class Cat(Pet):  # Inherits from Pet but replaces the speak() method
    def speak(self):  # Overrides the speak() method of Pet
        print(f"{self.name} says: Meow!")

# --- TYPE 4: Refinement (Cooperative) ---
# Logic: "I am a Pet. I'll let the Parent handle the name,
# but I will add 'barking' to the speak behavior."
class Dog(Pet):  # Inherits from Pet and refines the speak() method
    def __init__(self, name, weight):
        super().__init__(name)  # Let Pet handle the name
        self.weight = weight    # Dog handles the weight

    def speak(self):
        print(f"{self.name} lets out a loud growl...")
        super().speak()         # Then do the normal pet sound
        print(f"{self.name} barks!")

# --- Demonstration ---

print("--- Type 1: BasicPet ---")
p1 = BasicPet("Rocky")  # Inherits everything from Pet without changes
p1.speak()

print("--- Type 2: Retriever (Extension) ---")
p2 = Retriever("Buddy")  # Has additional fetch() method
p2.fetch() # New method

print("--- Type 3: Cat (Replacement) ---")
p3 = Cat("Kitty")  # Overrides speak() method
p3.speak() # Parent's generic sound is gone

print("--- Type 4: Dog (Refinement) ---")
p4 = Dog("Leo", 30)  # Adds weight attribute, and refines speak() method
p4.speak() # Combines Parent logic and Child logic
```

Example script for multiple inheritence

In this example, we study how multiple inheritance works in Python and how Python decides which method to execute when more than one parent class provides the same method.


```python
# A. BASE CLASS
class Pet:  # Base class for all pets
    def info(self):  # Common method for all pets
        print("I am a Pet")

# B. FIRST PARENT CLASS
class Walker(Pet):  # Walker inherits from Pet
    def action(self): # Override Pet's action() by implementing for Walker
        print("Walking... (Walker class)")

# C. SECOND PARENT CLASS
class Swimmer(Pet): # Swimmer inherits from Pet
    def action(self): # Override Pet's action() by implementing for Swimmer
        print("Swimming... (Swimmer class)")

# D. CHILD CLASS 1 (Walker first, then Swimmer). This will affect the MRO
class Dog1(Walker, Swimmer):  # Doesnt override action()
    pass  # So use the method from first parent in MRO (Walker)

# E. CHILD CLASS 2 (Swimmer first, then Walker). This will change the MRO
class Dog2(Swimmer, Walker):  #  Doesnt override action()
    pass  # So use the method from first parent in MRO (Swimmer)

# F. OBJECT CREATION
d1 = Dog1()  # Create Dog1 object
d2 = Dog2()  # Create Dog2 object

# G. METHOD CALLS
d1.action() # For Dog1 Walker comes first so Walker's action() will be used
# Output: Walking... (Walker class)

d2.action() # For Dog2 Swimmer comes first so Swimmer's action() will be used
# Output: Swimming... (Swimmer class)

# H. SHOW MRO
print("MRO of Dog1:", Dog1.__mro__)
# MRO for Dog1 is Dog1 → Walker → Swimmer → Pet → object
print("MRO of Dog2:", Dog2.__mro__)
# MRO for Dog2 is Dog2 → Swimmer → Walker → Pet → object
```

The following script shows a simple usage of composition:

```python

class Collar:  # Component class for composition
    def __init__(self, color):
        self.color = color

class Dog:  # Dog class that has a Collar (composition)
    def __init__(self, name):
        self.name = name
        self.collar = Collar("Red")   # Has-A relationship

d = Dog("Tommy")  # Create Dog object, which creates a Collar object inside it
# Accessing attributes of Dog and its composed Collar
print(d.name, d.collar.color)  # Output: Tommy Red
```

The following program demonstrates the concept of an abstract method using a base class.

```python
# A. BASE CLASS
class Pet:  # Base class for all pets
    def speak(self):  # Abstract method
        raise NotImplementedError("Subclass must implement speak()")

# B. CHILD CLASS 1
class Dog(Pet):  # Dog inherits from Pet
    def speak(self):  # Concrete method
        print("Dog barks")

# C. CHILD CLASS 2
class Cat(Pet):  # Cat inherits from Pet
    pass   # Does NOT implement speak()

# D. TESTING
d = Dog()  # Create Dog object
d.speak()   # Works

c = Cat()  # Create Cat object
# c.speak()   # NotImplementedError: Subclass must implement speak()
```

The following script shows the use of @abstractmethod from abc library:

```python
from abc import ABC, abstractmethod

class Pet(ABC):  # Abstract Base Class
    @abstractmethod
    def speak(self):
        pass

class Dog(Pet):
    def speak(self):
        print("Dog barks")

class Cat(Pet):  # Cat inherits from Pet
    pass   # ERROR at object creation

d = Dog()
d.speak()  # Dog barks

# c = Cat()   # TypeError Cannot create an object of Cat class
# because it does not implement the abstract method speak() from Pet.
# Advantage → early detection of error at object creation time,
# Unlike NotImplementedError → Late detection at runtime when method is called
```

Why Use Virtual Subclasses? (1) To define a common interface. (2) To allow unrelated classes to behave like a particular type. Useful in: Frameworks, Plugins and APIs
The script given below shows this


```python
from abc import ABC

# A. BASE CLASS (ABC)
class Pet(ABC):  # Abstract Base Class
    pass

# B. REAL SUBCLASS
class Dog(Pet):  # Dog inherits from Pet
    def speak(self):
        print("Dog barks")

# C. VIRTUAL SUBCLASS. Does NOT inherit from Pet
class RobotDog:  # But will be registered as virtual subclass
    def speak(self):
        print("Robot dog beeps")

# D. REGISTER RobotDog as virtual subclass of Pet without actual inheritance
Pet.register(RobotDog)
# RobotDog → is subclass of Pet for isinstance() and issubclass() checks
# But doesnt inherit from Pet so doesnt have access to methods/ attributes of Pet

# E. Create Objects
d = Dog()
r = RobotDog()

# F. Real subclass checks
print(issubclass(Dog, Pet))       # True. Dog is a real subclass of Pet.
print(isinstance(d, Pet))         # True. Dog is an instance of Pet.

# G. Virtual subclass checks
print(issubclass(RobotDog, Pet))  # True. RobotDog is a virtual subclass of Pet.
print(isinstance(r, Pet))         # True. RobotDog is an instance of Pet.

# H. MRO Check
print(Dog.__mro__)        # Shows inheritance
# print(RobotDog.__mro__) # Does NOT include Pet
```

The following script shows use of the __dict__ attribute to access, add or modify the attributes of an object.

```python
# A. CLASS DEFINITION
class Dog:
    species = "Pet"  # Class attribute

    def __init__(self, name):
        self.name = name  # Instance attribute

# B. OBJECT CREATION
d = Dog("Tommy")

# C. NAMESPACE DISPLAY
print("Class Namespace:->", Dog.__dict__)
# Output: Class Namespace:-> {'__module__': '__main__', 'species': 'Pet',....}
print("Instance Namespace:->", d.__dict__)
# Output: Instance Namespace:-> {'name': 'Tommy'}

# D. ADD NEW ATTRIBUTE by directly modifying the instance's __dict__
d.__dict__["color"] = "Brown"
print("Added attribute 'color':", d.color)  # Added attribute 'color': Brown

# E. MODIFY ATTRIBUTE by directly modifying the instance's __dict__
d.__dict__["name"] = "Bruno"  # Modifying the 'name' attribute using __dict__
print("Modified name:", d.name)  # Modified name: Bruno

```

The following script shows the effect of using and not using keywords global and nonlocal:

```python
# A. GLOBAL VARIABLES
pet_global_correct = "Global Dog (Original)"  # In global scope.
pet_global_wrong = "Global Cat (Original)"  # In global scope.

# B. FUNCTION: GLOBAL KEYWORD DEMO. Here pet_global_wrong is modified without
# using 'global', so it will not change the global variable. pet_global_correct
# is modified using 'global', so it will change the global variable.
def test_global():
    print("--- Inside test_global() ---")

    # WRONG: No 'global' keyword → creates LOCAL variable
    pet_global_wrong = "Modified without global (Wrong)"
    print("Inside function (wrong global):", pet_global_wrong)
    # Output: Inside function (wrong global): Modified without global (Wrong)

    # CORRECT: Using 'global' → modifies actual global variable
    global pet_global_correct
    pet_global_correct = "Modified with global (Correct)"
    print("Inside function (correct global):", pet_global_correct)
    # Output: Inside function (correct global): Modified with global (Correct)

# C. CALL FUNCTION
test_global()

# D. Check results outside function
print("--- After test_global() ---")
print("Global WRONG variable:", pet_global_wrong)     # Not changed
print("Global CORRECT variable:", pet_global_correct) # Changed

# E. FUNCTION: NONLOCAL KEYWORD DEMO. Here pet_enclosing_wrong is modified without
# using 'nonlocal', so it will not change enclosing variable. pet_enclosing_correct
# is modified using 'nonlocal' so it will change enclosing variable
def outer_function():
    print("--- Inside outer_function() ---")
    pet_enclosing_correct = "Enclosing Dog (Original)"
    pet_enclosing_wrong = "Enclosing Cat (Original)"

    def inner_function():
        print("--- Inside inner_function() ---")

        # WRONG: No 'nonlocal' → creates local variable
        pet_enclosing_wrong = "Modified without nonlocal (Wrong)"
        print("Inner (wrong nonlocal):", pet_enclosing_wrong)

        # CORRECT: Using 'nonlocal' → modifies enclosing variable
        nonlocal pet_enclosing_correct
        pet_enclosing_correct = "Modified with nonlocal (Correct)"
        print("Inner (correct nonlocal):", pet_enclosing_correct)

    # F. Call inner function. This will execute the code inside inner_function
    # and show effects of modifying variables with & without 'nonlocal' keyword
    inner_function()

    # G. Check results in enclosing scope
    print("--- Back in outer_function() ---")
    print("Enclosing WRONG variable:", pet_enclosing_wrong)     # Not changed
    print("Enclosing CORRECT variable:", pet_enclosing_correct) # Changed

# H. Call outer function
outer_function()
```

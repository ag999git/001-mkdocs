The following script shows the usage of some of the common functions of the random module:-



```python

import random

# Example 1:- random.random(). # It generates a random number
# in semi-open range [0.0, 1.0). The square bracket means
# "closed interval on the left" and the round parenthesis means
# "open interval on the right". So, 0.0 is possible, but all
# returned numbers # will be strictly less than 1.0.
random_float = random.random()
print(random_float) # Output: random float between 0.0 and 1.0

# Example 2:- random.randint(a, b). # It generates a
# random integer between a and b (both inclusive).
random_integer = random.randint(1, 10)
print(random_integer)  # Output: random integer between 1 and 10

# Example 3:- random.choice(sequence) It selects a random
# element from the given sequence (list, tuple, string, etc.)
my_list = ['East', 'West', 'North', 'South']
random_choice = random.choice(my_list)
print(random_choice)  # Output: A random element from the list

# Example 4:- random.shuffle(list). The list elements are
# shuffled in place changing their order randomly.
my_numbers = [1, 2, 3, 4, 5]
random.shuffle(my_numbers)
print(my_numbers)  # Output: shuffled list of numbers

# Example 5:- random.sample(population, k) returns a list
# of k unique elements chosen from the population sequence or set.
population = range(1, 21)  # Numbers from 1 to 20
sample_size = 5  # Number of elements to sample
random_sample = random.sample(population, sample_size)
print(random_sample) # Output: random sample of 5 numbers from population

# Example 6:- random.uniform(a, b) generates a
# random float number between a and b, including both endpoints a and b.
random_uniform = random.uniform(1.5, 5.5)
print(random_uniform)  # Output: random float between 1.5 and 5.5
```

**Note on `randrange([start[, stop, [step])`**

•	`start` -- Start point of the range. This would be included in the range.

•	`stop` -- Stop point of the range. This would be excluded from the range.

•	`step` -- Steps to be added in a number to decide a random number.
Return a randomly selected element from `range(start, stop, step)`.

Difference between `randrange([start[, stop, [step])` and `range(start, stop, step)` is `randrange([start[, stop, [step])` creates a range of integers. For example `range(3, 10,2)` -> will include `3,5,7,9`. But `randrange(3,10,2)` will give one of the integers `3,5,7,9`.

The following script shows the usage:-


```python
import random

# Example 1: Only stop value (10) given. Start defaults to 0
result = random.randrange(10)
# This gives a random integer from 0 to 9 (10 is not included)
print(result)  # can return ANY number from 0 to 9.
# Each time you run it, you may get a different result!

# Example 2: Both start (5) and stop (15) values given
result = random.randrange(5, 15)
print(result)  # can return ANY number from 5 to 14.
# Each time you run it, you may get a different result!

# Example 3: Start (0), stop (30), and step (5) values given
result = random.randrange(0, 30, 5)
print(result) # can return ANY number from 0 to 25 with
# a step of 5 (0, 5, 10, 15, 20, 25). But not 30, because stop value excluded
# Each time you run it, you may get a different result!

```

A function always returns something. Possible cases are: (1) If return value → returns the value. (2) If return alone → returns None. (3) If no return at all → returns None. The following script shows these 3 cases:-

```python
# Example 1: Function with a return value
def add_numbers(a, b):
    return a + b # The function returns the sum of a and b

result = add_numbers(5, 3)  # Function call
print("Result of add_numbers:", result)  # Output: 8

# Example 2: Function with return alone
def print_message():
    print("Hello, World!")
    return  # No value returned, implicitly returns None
message_result = print_message()  # Function call, Output: Hello, World!
print("Result of print_message:", message_result)  # Output: None

# Example 3: Function with no return statement
def greet(name):
    print(f"Hello, {name}!")  # Doesn't return anything, implicitly returns None
greet("Alice")  # Function call, Output: Hello, Alice!
greet_result = greet("Bob")  # Function call, Output: Hello, Bob!
print("Result of greet:", greet_result)  # Output: None
```

The following scripts shows how global and local are used:-

```python
# --- GLOBAL SCOPE ---
# Lifetime: Long (Exists as long as the program runs)
msg = "Hello"

def my_function():
    # --- LOCAL SCOPE ---
    # Lifetime: Short (Created when called, deleted when finished)
    name = "Alice"
    print(msg, name) # Can see both (Global and Local)

# --- CALL THE FUNCTION ---
my_function()  # Output: Hello Alice

# --- THE TEST ---
print(msg)     # Works! msg is Global.
# print(name)  # ERROR! name is Local and "died" when the function ended.
```

The following script shows name clash in local and global scope:-

```python
# --- GLOBAL SCOPE ---
# Global variable x exists in the "outer" world.
x = 'cat'

def f3():
    # --- LOCAL SCOPE ---
    # Shadowing: This creates a NEW local x.
    # It hides the global x while we are inside this function.
    x = 'dog'

    # Python checks the local "box" first, so it finds 'dog'.
    print('Inside f3() (Local)  ->', x)

# --- MAIN PROGRAM ---
f3() # This triggers the local variable's birth and death.

# Global x was never changed; it was just temporarily hidden.
print('Global scope (Main) ->', x)
```

**Creating two local scopes in a global scope.**

We can create a number of different local scopes within a global scope.

This is best understood by following example:-

```python
# Global Scope
x = 'cat in global scope' # This x belongs to the GLOBAL scope

def f1():
    # Local Scope of f1()
    x = 'dog in f1()' # Local x in f1() — shadows global x
    print("x in local scope of f1() ->", x)

    def f2():
        # Enclosed (Nested) Scope inside f1()
        x = 'rat in f2()'  # Local x in f2() shadows f1()'s x
        print("x in local scope of nested f2() ->", x)

    # Call nested function f2()
    f2()

# Function call creates temporary local scopes
f1()

# Local x inside f1() and inside f2() no longer exist here
print("x in global scope ->", x)

# Output:
# x in local scope of f1() -> dog in f1()
# x in local scope of nested f2() -> rat in f2()
# x in global scope -> cat in global scope
```

**Defining a function inside another function(Nested functions):-**


In python it is permitted to define one function inside another function (Also called nested functions). Following script shows this:-

```python
# Global Scope
x = 'cat in global scope' # This x belongs to the GLOBAL scope

def f1():
    # Local Scope of f1()
    x = 'dog in f1()' # Local x in f1() — shadows global x
    print("x in local scope of f1() ->", x)

    def f2():
        # Enclosed (Nested) Scope inside f1()
        x = 'rat in f2()'  # Local x in f2() shadows f1()'s x
        print("x in local scope of nested f2() ->", x)

    # Call nested function f2()
    f2()

# Function call creates temporary local scopes
f1()

# Local x inside f1() and inside f2() no longer exist here
print("x in global scope ->", x)

# Output:
# x in local scope of f1() -> dog in f1()
# x in local scope of nested f2() -> rat in f2()
# x in global scope -> cat in global scope
```

In the example given above, we were calling f2() from within f1().

You cannot call f2() directly in global scope.

You can do it if you "Return the nested function from f1()".This exposes the nested function to the global scope.

This is shown in the script below:-


```python
def f1():
    """Outer function that defines and returns another function."""

    def f2():
        """Inner function (Nested)."""
        print("Hello from f2!")

    # We return f2 without parentheses, not f2().
    return f2 # We are returning the function itself, not calling it.

# 1. Call f1(). It returns the function 'f2' and stores it in 'nested'.
# Calling f1() does not execute f2() but returns f2 as a function object.
nested = f1() # 'nested' holds f2 that was defined inside f1()

# 2. At this point, f1() is finished, but 'nested' holds f1's f2.
print(f"Type of 'nested' variable is: {type(nested)}")
# Output: Type of 'nested' variable is: <class 'function'>,
# confirming that 'nested' is indeed a function object.

# 3. We call the inner function from the global scope using the new name.
nested() # Output: Hello from f2!
# Explanation: 'nested' is now a reference to f2() defined inside f1().
```

Nested function and concept of closure

The following script explains the concept of closure


```python
def outer_greeter(name):  # 'name' is a parameter of the outer function.
    # This variable 'name' is in the Enclosing scope of inner_greeter
    # It is "captured" by the inner function below

    def inner_greeter():
        # Even after outer_greeter ends, this function
        # keeps 'name' alive so it can use it when called later.
        print(f"Hello, {name}!")

    return inner_greeter # Return the function object

# --- THE CAPTURE ---
# We create two different "instances" of the function
greet_alice = outer_greeter("Alice")
greet_bob = outer_greeter("Bob")

# --- THE EXECUTION ---
# The outer function 'outer_greeter' is long gone,
# but the inner functions still remember their specific data!
greet_alice()  # Output: Hello, Alice!
greet_bob()    # Output: Hello, Bob!
```

The following script shows how closure is used to "preserve state" without using "global variables". It does use a Python keyword "nonlocal"

```python
def make_counter():
    # 1. Variable count belongs to the ENCLOSING scope.
    # It acts as the "Private Memory" for the closure.
    count = 0

    def counter():
        # 2. The 'nonlocal' keyword is essential here.
        # It allows the inner function to MODIFY 'count'.
        # Without it, Python would treat 'count' as a new local variable.
        nonlocal count

        count += 1
        return count

    # 3. Return the function object to create the Closure
    return counter

# --- INDEPENDENT STATE ---
# Each call to make_counter() creates a brand new "Private Memory" box.
counter_a = make_counter()
counter_b = make_counter()

print(f"Counter A: {counter_a()}")  # Output: 1
print(f"Counter A: {counter_a()}")  # Output: 2

# Counter B starts fresh; it cannot see or touch Counter A's memory.
print(f"Counter B: {counter_b()}")  # Output: 1

# Counter A still remembers where it left off.
print(f"Counter A: {counter_a()}")  # Output: 3
```

Function composition is the process of combining two or more functions such that the output of one function becomes the input (argument) of the next.
The script given below demonstrates the difference in use of $f(g(x))$ and  $g(f(x))$ ,

```python
def f(x):
    """Adds 10 to the input"""
    return x + 10

def g(x):
    """Multiplies the input by 2"""
    return x * 2

# --- Case 1: f(g(x)) ---
# 1. g(5) runs first (5 * 2 = 10)
# 2. f(10) runs second (10 + 10 = 20)
result1 = f(g(5))  # Same as f(g(5)) which is f(10) which is 20

# --- Case 2: g(f(x)) ---
# 1. f(5) runs first (5 + 10 = 15)
# 2. g(15) runs second (15 * 2 = 30)
result2 = g(f(5))  # Same as g(f(5)) which is g(15) which is 30

print(f"f(g(5)) results in: {result1}") # Output: 20
print(f"g(f(5)) results in: {result2}") # Output: 30
```
